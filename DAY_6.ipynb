```python
import math

# 11 given customers (Age, Loan)
X = [
    [25, 40000],
    [35, 60000],
    [45, 80000],
    [20, 20000],
    [35, 120000],
    [52, 18000],
    [23, 95000],
    [40, 62000],
    [60, 100000],
    [48, 220000],
    [33, 150000]
]

# Their Default status (Yes / No)
Y = ['No','No','No','No','Yes','Yes','Yes','No','No','Yes','Yes']

# 12th customer (Age, Loan)
new = [48, 142000]

k = 3   # Number of nearest neighbors

# Step 1: Calculate distances
distances = []
for i in range(11):
    d = math.sqrt((new[0]-X[i][0])**2 + (new[1]-X[i][1])**2)
    distances.append([d, Y[i]])

# Step 2: Sort distances
distances.sort()

# Step 3: Take k nearest
nearest = distances[:k]

print("Nearest neighbors:", nearest)

# Step 4: Final decision
yes = 0
no = 0

for item in nearest:
    if item[1] == 'Yes':
        yes += 1
    else:
        no += 1

if yes > no:
    print("Prediction: 12th customer WILL DEFAULT")
else:
    print("Prediction: 12th customer will NOT DEFAULT")
```
